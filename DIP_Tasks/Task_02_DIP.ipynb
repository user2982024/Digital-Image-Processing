```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread(r"./2.jfif")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Convert to binary
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(binary, cmap="gray")
plt.title("Binary Image")
plt.axis("off")
```

```python
# connected components
num_labels, labels = cv2.connectedComponents(binary)
```

```python
total_objects = num_labels - 1
print("Total Number of Objects Detected =", total_objects)
```

```python
# create an empty color image
label_colored = np.zeros((labels.shape[0], labels.shape[1], 3))

for label in range(1, num_labels):  # skip background (0)
    label_colored[labels == label] = np.random.rand(3)  # random colors
```

```python
plt.imshow(label_colored)
plt.title("Connected Component Labeled Image")
plt.axis("off")
```

# Image Processing: Color Decomposition, Grayscale, Binary Conversion, and Connected Component Analysis

## 1. Introduction

In this task, we extend our basic image processing techniques to include **connected component analysis**, which allows us to detect and label distinct objects in a binary image. 

The steps performed in this task include:

1. **Loading and displaying the original colored image.**
2. **Color decomposition** into Red, Green, and Blue components.
3. **Grayscale conversion** to represent intensity information.
4. **Binary thresholding** to segment the image into foreground and background.
5. **Connected component labeling** to detect and assign unique labels to objects in the binary image.

This workflow demonstrates how simple transformations can lead to object detection and analysis, a fundamental step in **computer vision and image understanding**.

---

## 2. Methodology

### 2.1 Loading the Image
We begin by loading the image using **OpenCV**. OpenCV reads images in **BGR format**, which is converted to **RGB** for proper visualization using matplotlib.

### 2.2 RGB Component Decomposition
To understand the contribution of each color channel, we create separate images highlighting only:

- Red Component (set Green and Blue to zero)
- Green Component (set Red and Blue to zero)
- Blue Component (set Red and Green to zero)

This is helpful for:

- Color-based segmentation
- Feature extraction
- Detecting objects or regions with dominant colors

### 2.3 Grayscale Conversion
The RGB image is converted to **grayscale**, reducing color information to intensity values. This simplifies further processing, especially thresholding and object detection.

### 2.4 Binary Thresholding
A global threshold (127) is applied to convert the grayscale image to **binary**. Pixels above the threshold become 255 (white, foreground), and below become 0 (black, background). Binary images are essential for:

- Object segmentation
- Counting discrete objects
- Connected component analysis

### 2.5 Connected Component Labeling
Connected component analysis labels **each distinct object** in the binary image. Steps:

1. Identify connected pixels with value 255.
2. Assign a unique label to each object.
3. Generate a **colored labeled image**, where each object is shown in a unique color.

This process enables **object counting and visualization**.

---

## 3. Observations

- The **original image** shows all color information.
- The **Red, Green, and Blue component images** reveal regions dominated by each color.
- The **grayscale image** emphasizes intensity variations, independent of color.
- The **binary image** segments objects from the background.
- Connected component labeling identifies all discrete objects and assigns **unique colors** for visualization.
- The **total number of objects detected** can be printed and analyzed.
