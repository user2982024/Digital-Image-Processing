```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

image_path = '5.jfif'

try:
    # Open the original image
    img = Image.open(image_path)

    # Apply filters
    filtered_img_mean = img.filter(ImageFilter.BoxBlur(3))
    filtered_img_median = img.filter(ImageFilter.MedianFilter(size=3))
    filtered_img_mode = img.filter(ImageFilter.ModeFilter(size=3))
    filtered_img_gaussian = img.filter(ImageFilter.GaussianBlur(radius=2))

    # Create a figure with multiple subplots
    plt.figure(figsize=(18, 12))

    # Original Image
    plt.subplot(2, 3, 1) # 2 rows, 3 columns, 1st plot
    plt.imshow(img)
    plt.title('Original Image')
    plt.axis('off')

    # Mean Filtered Image
    plt.subplot(2, 3, 2) # 2 rows, 3 columns, 2nd plot
    plt.imshow(filtered_img_mean)
    plt.title('Mean Filter (BoxBlur Radius 3)')
    plt.axis('off')

    # Median Filtered Image
    plt.subplot(2, 3, 3) # 2 rows, 3 columns, 3rd plot
    plt.imshow(filtered_img_median)
    plt.title('Median Filter (Size 3)')
    plt.axis('off')

    # Mode Filtered Image
    plt.subplot(2, 3, 4) # 2 rows, 3 columns, 4th plot
    plt.imshow(filtered_img_mode)
    plt.title('Mode Filter (Size 3)')
    plt.axis('off')

    # Gaussian Filtered Image
    plt.subplot(2, 3, 5) # 2 rows, 3 columns, 5th plot
    plt.imshow(filtered_img_gaussian)
    plt.title('Gaussian Filter (Radius 2)')
    plt.axis('off')

    plt.tight_layout() # Adjust subplot parameters for a tight layout
    plt.show()

except FileNotFoundError:
    print(f"Error: The file at {image_path} was not found. Please check the path.")
except Exception as e:
    print(f"An error occurred: {e}")
```

# Image Smoothing and Filtering Techniques

## 1. Introduction

Image smoothing and filtering are fundamental techniques in image processing. They are used to:

- Reduce noise in images
- Remove small artifacts
- Highlight important features
- Prepare images for further processing like edge detection or segmentation

In this task, we apply **different spatial filtering techniques** using the Python Imaging Library (PIL):

1. **Mean Filter (Box Blur)**
2. **Median Filter**
3. **Mode Filter**
4. **Gaussian Blur**

---

## 2. Methodology

### 2.1 Loading the Image
- The image is loaded using `PIL.Image.open()`.
- The image is displayed using `matplotlib.pyplot`.

### 2.2 Mean Filter (Box Blur)
- Smooths the image by replacing each pixel with the **average value of its neighborhood**.
- Reduces noise but may blur edges slightly.
- Implemented using `ImageFilter.BoxBlur(radius)`.

### 2.3 Median Filter
- Replaces each pixel with the **median value of its neighborhood**.
- Very effective at removing **salt-and-pepper noise**.
- Implemented using `ImageFilter.MedianFilter(size)`.

### 2.4 Mode Filter
- Replaces each pixel with the **most frequent value (mode)** in its neighborhood.
- Useful for **categorical images** or for noise reduction without blurring edges.
- Implemented using `ImageFilter.ModeFilter(size)`.

### 2.5 Gaussian Filter
- Applies a **weighted average** where central pixels are given more weight.
- Smooths the image while **preserving edges better** than the mean filter.
- Implemented using `ImageFilter.GaussianBlur(radius)`.

---

## 3. Visualization
- The **original image** and all **filtered versions** are displayed side by side.
- This allows a clear **comparison of the effects** of each filtering technique.

---

## 4. Observations
- **Mean Filter**: Smooths overall appearance, edges slightly blurred.
- **Median Filter**: Excellent for removing sparse noise, edges preserved better than mean filter.
- **Mode Filter**: Reduces repetitive noise, preserves edges, good for low-texture areas.
- **Gaussian Filter**: Smooths image while maintaining edge structure; preferred for natural images.

---

## 5. Conclusion
- Different filtering techniques serve different purposes.
- **Choice of filter** depends on:
  - Type of noise
  - Importance of edge preservation
  - Desired visual quality
- Combining filters can further improve noise removal and enhancement for preprocessing in computer vision tasks.
