Step 1: Install & import libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Step 2: Load image

```python
# Read image
img = cv2.imread('8.jfif')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Convert to binary
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(binary, cmap='gray')
plt.title("Original Binary Image")
plt.axis('off')
```

Structuring element

```python
kernel = np.ones((3,3), np.uint8)
```

1. Erosion

```python
erosion = cv2.erode(binary, kernel, iterations=1)

plt.imshow(erosion, cmap='gray')
plt.title("Erosion")
plt.axis('off')
```

2. Dilation

```python
dilation = cv2.dilate(binary, kernel, iterations=1)

plt.imshow(dilation, cmap='gray')
plt.title("Dilation")
plt.axis('off')
```

3. Opening

```python
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)

plt.imshow(opening, cmap='gray')
plt.title("Opening")
plt.axis('off')
```

4. Closing

```python
closing = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)

plt.imshow(closing, cmap='gray')
plt.title("Closing")
plt.axis('off')
```

2. Boundary Extraction

```python
boundary = binary - erosion

plt.imshow(boundary, cmap='gray')
plt.title("Boundary Extraction")
plt.axis('off')
```

3. Fill Holes using Reconstruction

```python
filled = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel, iterations=3)

plt.imshow(filled, cmap='gray')
plt.title("Hole Filled Image")
plt.axis('off')
```

4. Remove Noise using Morphology

```python
noise_removed = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
noise_removed = cv2.morphologyEx(noise_removed, cv2.MORPH_CLOSE, kernel)

plt.imshow(noise_removed, cmap='gray')
plt.title("Noise Removed Image")
plt.axis('off')
```

5. Detect Specific Shapes

```python
kernel_square = cv2.getStructuringElement(cv2.MORPH_RECT, (15,15))

shape_detect = cv2.erode(binary, kernel_square, iterations=1)

plt.imshow(shape_detect, cmap='gray')
plt.title("Detected Shape")
plt.axis('off')
```

# Morphological Operations on Binary Images

**Objective:**  
Perform **morphological image processing** techniques on a binary image to analyze structures, remove noise, fill holes, and detect shapes.

---

### Step 1: Read and Binarize Image

- Load the input image.
- Convert it to **grayscale** and then to **binary** using a threshold.
- Binary image is required for most morphological operations.

**Outcome:**  
- Original binary image ready for morphological processing.

---

### Step 2: Erosion

- **Erosion** removes pixels on object boundaries.
- Uses a structuring element (kernel) to shrink objects.
- Useful to remove small white noise or detach connected objects.

**Outcome:**  
- Objects in the image shrink.
- Boundaries move inward.

---

### Step 3: Dilation

- **Dilation** adds pixels to object boundaries.
- Expands shapes and fills small holes.
- Useful to emphasize features or connect objects.

**Outcome:**  
- Objects grow.
- Boundaries move outward.

---

### Step 4: Opening and Closing

- **Opening**: Erosion followed by dilation.
  - Removes small objects or noise without affecting the overall shape.
- **Closing**: Dilation followed by erosion.
  - Fills small holes and gaps in objects.

**Outcome:**  
- Opening → Noise removed.  
- Closing → Holes filled.

---

### Step 5: Boundary Extraction

- Extract boundaries by subtracting the eroded image from the original.
- Highlights the **edges of objects**.

**Outcome:**  
- Binary image showing object boundaries.

---

### Step 6: Hole Filling

- Apply multiple closing operations to fill holes in objects.
- Useful for preparing images for shape analysis or feature extraction.

**Outcome:**  
- Objects become solid without internal gaps.

---

### Step 7: Noise Removal

- Apply **opening followed by closing** to remove small noise while preserving object shapes.
- Helps in cleaning the binary image for further processing.

**Outcome:**  
- Noise-free binary image suitable for analysis.

---

### Step 8: Shape Detection

- Use a **larger square structuring element** for erosion to detect significant shapes.
- Only objects larger than the structuring element remain after erosion.

**Outcome:**  
- Detected major shapes in the image.
- Small or irrelevant features are removed.

---

### Summary:

- Morphological operations manipulate the **structure of objects** in binary images.
- Key operations: **Erosion, Dilation, Opening, Closing, Boundary Extraction, Hole Filling, Noise Removal, Shape Detection**.
- Applications include:
  - Preprocessing for segmentation.
  - Feature extraction.
  - Object detection and analysis.
- Visualization of each step helps understand the impact of these operations on the binary image.
